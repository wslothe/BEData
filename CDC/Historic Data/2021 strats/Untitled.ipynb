```r
library(dplyr)
library(epitools)
file_path <- "numeric_cases_21.csv"
cases21 <- read.csv(file_path)
adults <- cases21[!cases21$age_group == "0 - 17 years", ]
df <- adults[!adults$age_group == " ", ]
df_complete <- df[!df$symp == "", ]
dataframe = table(df_complete$age_group, df_complete$symp)[-1:-2,]
dataframe
```

```r
library(epitools)
oddsratio(dataframe, correction = TRUE)
```

```r
riskratio.wald(dataframe, rev=c("rows"))
```

```r
data <- df[!df$hosp == "", ]
agehosp = table(data$age_group, data$hosp)[-1:-2,]
agehosp
```

```r
oddsratio.wald(agehosp)
```

```r
riskratio.wald(agehosp)
```

```r
icu <- df[!df$icu == "", ]
ageicu = table(data$age_group, data$icu)[-1:-2,]
```

```r
oddsratio.wald(ageicu)
```

```r
rrageicu = riskratio.small(ageicu, correction = TRUE)
rrageicu
```
